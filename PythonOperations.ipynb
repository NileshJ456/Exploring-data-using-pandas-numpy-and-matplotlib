```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load data into pandas dataframe
df= pd.read_csv('Data.csv')
df
```

```python
#Removing salary and salary_currency columns from dataframe
df1= df.drop(["salary","salary_currency"],axis='columns')
df1
```

```python
#Drop rows that contain Nan values from updated dataframe
df2 = df1.dropna()
df2
```

```python
#To check duplicates in the dataframe
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

```python
# To see unique values from all the columns in dataframe
for col in df.columns:
    print(f"Unique values in column {col}: {df[col].unique()}")
```

```python
# Define the mapping for renaming experience levels
experience_level_mapping = {'SE': 'Senior Level/Expert', 'MI': 'Medium Level/Intermediate', 'EN': 'Entry Level', 'EX': 'Executive Level'}

# Rename the 'experience_level' column
df['experience_level'] = df['experience_level'].replace(experience_level_mapping)

# Print the dataFrame to verify the changes
print(df)
```

```python
#Initialize the value of total salary to 0
total_salary= 0
#For loop to iterate through the loop
for salary in df.iloc[:,4]:
    total_salary += salary
print("Total salary:", total_salary)
```

```python
# Get the number of rows
num_rows = df.shape[0]

#Initialize the value of mean to 0
mean= 0
#For loop to iterate through loop
for salary in df.iloc[:,4]:
    mean= total_salary / num_rows
print(f"Mean: {mean:.2f}")
```

```python
# Calculate standard deviation of salary
standard_deviation = np.std(df['salary'])

# Print the result
print(f"Standard Deviation: {standard_deviation:.2f}")
```

```python
# Calculate skewness and kurtosis
skewness = df['salary'].skew()
kurtosis = df['salary'].kurt()

#Print results
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
# Select only numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation of all variables with numerical values
correlation = numerical_columns.corr()

# Print the correlation matrix of all variables with numerical values
print(correlation)
```

```python
# Find the top 15 jobs
top_jobs = df['job_title'].value_counts().head(15)

# Create a bar graph of top 15 jobs
plt.figure(figsize=(10, 6))
top_jobs.plot(kind='bar')
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Find the job with the highest salary
highest_salary_job = df.loc[df['salary_in_usd'].idxmax(), 'job_title']

# Create a bar graph to illustrate the highest salary job
plt.figure(figsize=(8, 6))
plt.bar(highest_salary_job, df['salary_in_usd'].max(), color='lightgreen')
plt.title('Job with the Highest Salary')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by experience level and calculate average salary for each level
avg_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean()

# Create a bar graph to illustrate average salaries based on experience level
plt.figure(figsize=(10, 6))
avg_salary_by_experience.plot(kind='bar', color='orange')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Plotting histogram of experience_level
plt.figure(figsize=(8, 6))
plt.hist(df['experience_level'], bins=len(df['experience_level'].unique()), color='skyblue', alpha=0.7)
plt.title('Histogram of Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
# Plotting box plot of salary
plt.figure(figsize=(10, 6))
plt.boxplot(df['salary'], vert=False)
plt.title('Box Plot of Salary')
plt.xlabel('Salary (USD)')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
